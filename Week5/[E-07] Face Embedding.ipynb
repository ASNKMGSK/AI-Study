## 7.2 얼굴 영역 자르기

```python
# jpg형식의 이미지를 불러온 후 얼굴을 face_recognition.face_locations를 사용해 찾아준다
import face_recognition
import os
import matplotlib.pyplot as plt

image_path = os.getenv('HOME')+'/aiffel/face_embedding2/images/강재현.jpeg'
image = face_recognition.load_image_file(image_path)
face_locations = face_recognition.face_locations(image)

print(face_locations)  # 이미지에서 얼굴 영역의 좌표를 출력

%matplotlib inline

a, b, c, d = face_locations[0]
cropped_face = image[a:c,d:b,:]

plt.imshow(cropped_face)   # 이미지에서 얼굴영역만 잘라낸 cropped_face를 그려보자
```

```python
# 이미지 파일 경로를 파라미터로 넘기면 얼굴 영역만 잘라서 뿌려주는 함수를 하나 만들어 준다
import face_recognition
import os

def get_gropped_face(image_file):
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    a, b, c, d = face_locations[0]
    cropped_face = image[a:c,d:b,:]

    return cropped_face
```

```python
# 위에서 만든 함수가 잘 만들어졌는지 다른 사진을 통해 확인해보자
image_path = os.getenv('HOME')+'/aiffel/face_embedding2/images/actor35.jpeg'

cropped_face = get_gropped_face(image_path)
plt.imshow(cropped_face)
```

## 7.3 얼굴 영역의 임베딩 추출하고 비교하기

```python
# 강동원.jpeg에 포함된 얼굴 영역의 임베딩 벡터를 구해보자. 이전 스텝에서 이미지 파일의
# 얼굴 영역을 가져오는 get_gropped_face() 함수를 구현했으니 그것을 활용한다.
image_file = os.path.join(dir_path, 'actor29.jpeg')
face = get_gropped_face(image_file)   # 얼굴 영역을 구하는 함수(이전 스텝에서 구현)

# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)  
embedding
```

```python
# images 디렉토리 안에 있는 모든 이미지 파일의 임베딩을 구해서 dict 구조에 담아 리턴하는 함수를 하나 만든다
# 딕셔너리의 각 엘리먼트는 "사람 이름: 임베딩 벡터"로 한다
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}

    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_gropped_face(img_path)
        embedding = get_face_embedding(face)
        if len(embedding) > 0:  # 얼굴영역 face가 제대로 detect되지 않으면  len(embedding)==0인 경우가 발생하므로 
                    # os.path.splitext(file)[0]에는 이미지파일명에서 확장자를 제거한 이름이 담깁니다. 
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]

    return embedding_dict
```

```python
# 위에서 만든 함수가 잘 만들어졌는지 확인해 보자
embedding_dict = get_face_embedding_dict(dir_path)
embedding_dict['actor35']
```

```python
# 위의 단계에서 이미지의 얼굴 임베딩을 구해두었으니 다른 2개의 사진 임베딩 벡터 간의 거리를 계산하는 함수를 구현해보자
# get_distance() 함수는 인자로 파일명의 사람 이름 2개를 받아서, 두 사람의 얼굴 임베딩 사이의 거리(L2 distance)를 구하는 형태다
import numpy as np
def get_distance(name1, name2):
    return np.linalg.norm(embedding_dict[name1]-embedding_dict[name2], ord=2)

get_distance('강재현', 'actor29')
```

```python
# name1과 name2의 거리를 비교하는 함수를 생성하되, name1은 미리 지정하고, name2는 호출시에 인자로 받도록 한다
def get_sort_key_func(name1):
    def get_distance_from_name1(name2):
        return get_distance(name1, name2)
    return get_distance_from_name1

sort_key_func = get_sort_key_func('actor31')   
# 이렇게 생성된 함수 sort_key_func는 sort_key_func('강재현') 라고 호출할 때 '박형식'과 '강재현' 사이의 임베딩 벡터 거리를 계산한다
```

```python
# 우리는 이미 모든 파일에 대한 얼굴 임베딩 딕셔너리를 가지고 있다.
# 이제 우리는 이 딕셔너리를 오름차순으로 정렬하되, 정렬 기준을 바로 '강재현'과의 임베딩 벡터 거리 함수로 할 것이다.
# 이때 lamda 함수가 정렬 key로 활용된다.
sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))
```
